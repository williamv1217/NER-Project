```python
from nltk.corpus import brown
from nltk.util import ngrams
from collections import Counter
```

```python
unigrams = ngrams(brown.words(), 1)
unigram_freq = Counter(unigrams)
```

```python
bigrams = ngrams(brown.words(), 2)
bigram_freq = Counter(bigrams)

```

```python
trigrams = ngrams(brown.words(), 3)
trigram_freq = Counter(trigrams)

```

```python
print('unigram total: ', len(unigram_freq))
print('bigram total: ', len(bigram_freq))
print('trigram total: ', len(trigram_freq))
```

```python
print(bigram_freq.most_common(5))
```

```python
print('the President: ', bigram_freq.get(('the', 'President')))
print('the Russian: ', bigram_freq.get(('the', 'Russian')))
print('boiled haddock: ', bigram_freq.get(('boiled', 'haddock')))
```

```python
from operator import itemgetter
ds1 = set()
for x in brown.words():
    tf = trigram_freq.get(('ran', 'the', x))
    if tf is not None:
        ds1.add((x, tf))
#         print(f"('ran', 'the', {x}) --- {tf}")
# print(ds1)
print(sorted(ds1, key=itemgetter(1), reverse=True)[:5])
```

```python
ds2 = set()
for x in brown.words():
    bf = bigram_freq.get((x, 'drinks'))
    if bf is not None:
        ds2.add((x, bf))
print(sorted(ds2, key=itemgetter(1), reverse=True)[:5])
```

```python
ds3 = set()
for x in brown.words():
    df = trigram_freq.get(('in', 'the', x))
    if df is not None:
        ds3.add((x, df))
print(sorted(ds3, key=itemgetter(1), reverse=True)[:5])
```

```python
for x in bigram_freq:
    print(f'{x[0]} {x[1]}\t{bigram_freq[x]}')
```

```python
with open('unigram_file.txt', 'w') as f:
    for x in unigram_freq:
        f.write(f'{x[0]}\t{unigram_freq[x]}\n')
f.close()
```

```python
with open('bigram_file.txt', 'w') as f:
    for x in bigram_freq:
        f.write(f'{x[0]} {x[1]}\t{bigram_freq[x]}\n')
f.close()
```

```python
with open('trigram_file.txt', 'w') as f:
    for x in trigram_freq:
        f.write(f'{x[0]} {x[1]} {x[2]}\t{trigram_freq[x]}\n')
f.close()
```
